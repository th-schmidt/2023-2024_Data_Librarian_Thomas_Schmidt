### Aufgabenbeschreibung
Wir möchten nun zusätzlich zu den Bibsonomy Daten auch noch Daten aus einer weiteren Quelle zusammenführen, in diesem Fall Semantic Scholar.

1. Lade "bibsonomy.json" und "semantic_scholar.json" in Pandas als ein Dataframe, füge eine Spalte "source" hinzu, in der du festhälst, woher die Suchergebnisse kommen.
2. Selektiere und benenne die Spalten so um dass aus beiden Quellen Titel, Autoren, URL, Abstract, Keywords, Query ( Suchanfrage) und das Jahr enthalten und gleich benannt sind.
3. Führe die Datensätze zusammen.
4. Stelle sicher, dass alle Einträge für "Jahr" eine 4-stellige Zahl (int) ist.
5. Stelle sicher, dass die Autoren gleich als Liste formatiert sind (also z. B. [Max Mustermann, Lieschen Müller] oder [Mustermann M., Müller L.]).
6. Entferne Duplikate (anhand der Spalte Title).
7. Sorge dafür, dass es einen durchgängigen Index gibt.

```python
import pandas as pd
```

```python
df = pd.read_json('./bibsonomy.json')
df['source'] = 'bibsonomy'
df.head(2)
```

```python
df.columns
```

```python
df2 = pd.read_json('./semantic_scholar.json')
df2['source'] = 'semantic_scholar'
df2.head(2)
```

```python
df2.columns
```

## Spaltenauswahl festlegen

```python
relevant_cols_df1 = ['title', 'author', 'year', 'url', 'abstract', 'tags', 'query', 'source']

df_filtered = df.filter(relevant_cols_df1)
df_filtered.head(2)
```

```python
relevant_cols_df2 = ['title', 'authors', 'year', 'url', 'abstract', 'fieldsOfStudy', 'searchQuery', 'source']

df2_filtered = df2.filter(relevant_cols_df2)
df2_filtered = df2_filtered.rename(columns={'authors': 'author', 'searchQuery': 'query', 'fieldsOfStudy': 'tags'})
df2_filtered.head(2)
```

## Datensätze zusammenführen

```python
df_complete = pd.concat([df_filtered, df2_filtered])
df_complete.head()
```

## Clean `author` column

```python
df_complete[df_complete['source'] == 'semantic_scholar'][['author']]
```

```python
def clean_authors(author):
    full_names = []
    for x in author:
        if isinstance(x, dict): # semantic_scholar
            full_names.append(x['name'])
        else:
            full_names.append(x) # bibsonomy
    return full_names

df_complete['clean_author'] = df_complete['author'].apply(clean_authors)
```

```python
df_complete[df_complete['source'] == 'semantic_scholar'].head(3)
```

## Clean `year` column

```python
df_complete['year'] = df_complete['year'].str.replace('\D+', '', regex=True)
df_complete.head(2)
```

```python
df_complete = df_complete.dropna(subset='year')
```

```python
df_complete['year'] = df_complete['year'].astype('int32')
df_complete['year'].dtype
```

## Duplikate entfernen

```python
len(df_complete)
```

```python
df_complete = df_complete.drop_duplicates(subset='title')
len(df_complete)
```

## Neuer Index

```python
df_final = df_complete.reset_index(drop=True)
df_final
```
