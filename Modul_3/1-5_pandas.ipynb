```python
import pandas as pd
```

```python
# create a dataframe by hand with two columns and three rows
df = pd.DataFrame({
        'month': [1, 2, 3],
        'temperatur': [-12, 3, 9]
})

# print out some descriptive statistics
df.describe()
```

```python
new_data = {
    'month': [4, 5, 6],
    'temperatur': [11, 19, 22]
}
new_data
```

```python
df = pd.concat([df, pd.DataFrame(new_data)], ignore_index=True)
df
```
