## Groupby

```python
apcs_publishers_and_spending = apcs[['publisher', 'euro']]
apcs_publishers_and_spending.head()
```

```python
apcs_publishers_and_spending.shape
```

```python
apcs_publishers_and_spending.describe()
```

```python
apcs_publishers_and_spending.groupby('publisher')
```

```python
apcs_publishers_and_spending.groupby('publisher').sum().sort_values('euro', ascending=False)
```

```python
apcs_publishers_and_spending.groupby('publisher').median().sort_values('euro', ascending=False)
```

```python
apcs_publishers_and_spending.groupby('publisher').sum().sort_values('euro', ascending=True).plot.barh();
```
